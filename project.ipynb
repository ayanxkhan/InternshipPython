```python
amazon_Dataset=pd.read_csv("7817_1.csv")
amazon_reviews=amazon_Dataset[['reviews.text']]
print(amazon_reviews)
```

```python
trivial_values = ['?', '-', ' ', 'nan', None]
amazon_reviews = amazon_reviews.replace(trivial_values, pd.NA).dropna()
print(amazon_reviews.shape)
print(amazon_reviews.head())
```

```python
amazon_ratings=amazon_Dataset[['reviews.rating']]
print(amazon_ratings.head())
amazon_Dataset['Sentiment'] = amazon_Dataset['reviews.rating'].apply(
    lambda x: 'Positive' if x >= 4 else ('Negative' if x <= 2 else 'Neutral')
)
print(amazon_Dataset['Sentiment'])
```

```python
print(imdb_Dataset.head())
print(amazon_Dataset.columns)
```

```python
amazon_Dataset = amazon_Dataset[['reviews.text', 'Sentiment']]
amazon_Dataset.columns = ['review', 'sentiment']



```

```python
imdb_Dataset = imdb_Dataset[['review', 'sentiment']]
imdb_Dataset.shape
amazon_Dataset.shape
```

```python
%pip install vaderSentiment

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()

def get_sentiment(text):
    score = analyzer.polarity_scores(text)['compound']
    return 'Positive' if score >= 0.05 else 'Negative' if score <= -0.05 else 'Neutral'

twitter_Dataset['sentiment'] = twitter_reviews.apply(get_sentiment)
print(twitter_Dataset['sentiment'])
```

```python
print(twitter_Dataset.columns)
# Rename the last two columns
twitter_df = twitter_Dataset.iloc[:, [-2, -1]]
twitter_df.columns = ['review', 'sentiment'] 
# Rename for consistency
twitter_df.shape
```

```python
combined_df = pd.concat([amazon_Dataset, imdb_Dataset, twitter_df], ignore_index=True)

#print(final_Dataset.head())
print(combined_df.columns)
combined_df.shape
```

```python
print(combined_df.head())
```

```python
import re

# Drop NaN values
combined_df = combined_df.dropna(subset=['review', 'sentiment'])

# Convert text to lowercase
combined_df['review'] = combined_df['review'].str.lower()

# Remove URLs
combined_df['review'] = combined_df['review'].apply(lambda x: re.sub(r'http\S+|www.\S+', '', x))

# Remove special characters and numbers
combined_df['review'] = combined_df['review'].apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', x))

# Remove extra whitespaces
combined_df['review'] = combined_df['review'].apply(lambda x: " ".join(x.split()))

print(combined_df.head())
```

```python
#sentiment distribution
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x=combined_df['sentiment'])
plt.title("Sentiment Distribution")
plt.show()
#mistake their are 5 columns because was uppercase and inconsistency
```

```python
# Standardize sentiment labels

combined_df['sentiment'] = combined_df['sentiment'].str.strip().str.capitalize()

# Check unique sentiment labels after fixing
print(combined_df['sentiment'].unique())
```
